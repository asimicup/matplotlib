```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
import random
```

```python
ml_std_age = np.random.randint(18, 45, (100))
py_std_age = np.random.randint(15, 40, (100))

print(ml_std_age)
print(py_std_age)
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(ml_std_age, bins, color = 'g', rwidth = 0.8, histtype = 'bar', label = 'ML std')

plt.hist(py_std_age, bins, color = 'r', rwidth = 0.8, histtype = 'bar', label = 'Py std')

plt.title("Machine Learning Students")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")
plt.legend()
```

```python
style.use('ggplot')
plt.figure(figsize = (10,8))
plt.hist([ml_std_age, py_std_age], bins, color = ['g', 'm'], rwidth = 0.8, histtype = 'bar', label = ['ML Std', 'Py Std'])

# plt.hist(py_std_age, bins, color = 'r', rwidth = 0.8, histtype = 'bar', label = 'Py std')

plt.title("Machine Learning Students")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")
plt.legend()
```
