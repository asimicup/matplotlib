```python
df1 = pd.read_csv('D:\\Projects\\ML Practice\\3. Matplotlib\\google play store apps\\googleplaystore.csv', nrows = 1000)
df1.shape
```

```python
X = df1['Rating']
Y = df1['Reviews']
```

```python
plt.figure(figsize = (10,8))

plt.scatter(X, Y, c = 'r', marker = '*', s = 100, linewidth = 2)
plt.scatter(X, df1['Installs'], c = 'g', marker = '*', s = 100, linewidth = 2)

plt.title('Google Play Store Apps')
plt.xlabel('Ratings')
plt.ylabel('Reviews and Installs')
```
