```python
import matplotlib.pyplot as plt
```

```python
classes = ['Python', 'AI', 'C++', 'ML', 'Data Science']

cls1_std = [40, 15, 30, 25, 40]

explode = [0,0,0.1,0,0]
colors = ['c', 'b', 'r', 'y', 'g']

plt.pie(cls1_std, labels = classes, explode = explode, colors = colors, autopct = '%0.2f%%',
       shadow = True, radius = 1.8, startangle = 90, )
plt.show
```
