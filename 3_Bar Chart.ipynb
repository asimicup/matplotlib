```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
```

```python
classes = ['Python', 'R', 'DS', 'ML', 'AI']

cls1_std = [30, 10, 20, 25, 10]
cls2_std = [40, 5, 20, 20, 10]
cls3_std = [35, 5, 30, 15, 15]
```

```python
style.use('ggplot')
plt.figure(figsize = (8, 5))

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, cls1_std, width, label = 'Class 1 Students', 
        align = 'edge', color = 'b', edgecolor = 'm', linewidth = '3')

plt.bar(classes_index  + width, cls2_std, width, label = 'Class 1 Students', 
        align = 'edge', color = 'g', edgecolor = 'm', linewidth = '3')

plt.bar(classes_index  + width + width, cls3_std, width, label = 'Class 1 Students', 
        align = 'edge', color = 'y', edgecolor = 'm', linewidth = '3')

plt.xticks(classes_index + width, classes)

plt.title('Bar Chart of Classes')
plt.xlabel('Classes')
plt.ylabel('No. of students')
plt.legend()
```
